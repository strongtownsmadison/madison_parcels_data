```python
import geopandas as gpd
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
alders = gpd.read_file('Alder_Districts.geojson')
lakes_rivers = gpd.read_file('Lakes_and_Rivers.geojson')
```

```python
res_symdiff = alders.overlay(lakes_rivers, how='symmetric_difference')
res_symdiff.plot()
```

```python
res_difference = alders.overlay(lakes_rivers, how='difference')
res_difference.plot()
```
